```python
import re
import collections

import pandas as pd
```

```python
df_2020_emp = pd.read_excel('AY 2020-BBE-EMP Free Response.xlsx',sheet_name=3, skiprows=1, index_col=0) 
df_2021_emp = pd.read_excel('AY 2021 BBE EFR Free Response.xlsx',sheet_name=3, skiprows=1, index_col=0) 
```

```python
df_2020_emp['year'] = "2020"
df_2021_emp['year'] = "2021"
```

```python
df = pd.concat([df_2020_emp,df_2021_emp])
df = pd.DataFrame(df)
df.head(2)
```

```python
df = df.rename(columns={
    "Accurately define a problem or issue in the workplace": "Problem Definition",
    "Assess advantages and disadvantages of each solution/course of action": "Analyze pros and cons of solutions",
    "Assess and determine if the problem has been resolved by the new course of action":"Issue resolution",
    "Identify all potential causes of the problem":"Identify all potential causes",
    "Identify the primary cause of the problem/issue":"Root cause identification",
    "Implement continuous changes for improvement if necessary":"Continuous improvements",
    "Implement most effective solution to resolve identified problem/issue":"Execute the optimal fix for the problem"
})
df.head(2)
```

```python
columns_to_drop = ['College', 'Work Terms','Class BOT','Citizenship Status','Co-op #', 'Coll', 'Conc 1', 'Coop #','Class BOT','Citizenship','Terms']
```

```python
df.drop(columns=columns_to_drop, inplace=True)
df.head(2)
```

```python
selected_columns = ['Problem Definition', 'Analyze pros and cons of solutions', 'Issue resolution',
                    'Identify all potential causes', 'Root cause identification',
                    'Continuous improvements',
                    'Execute the optimal fix for the problem']


avg_scores = df.groupby(['Major', 'year'])[selected_columns].mean().reset_index()


avg_scores.to_excel('average_scores_selected_columns.xlsx', index=False)
```
